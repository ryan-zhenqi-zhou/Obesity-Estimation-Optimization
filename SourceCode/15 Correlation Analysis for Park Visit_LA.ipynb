```python
# Standardization function
def standarize_data(data, stats):
    return (data - stats['mean'])/ stats['std']

df1_stats = df1.describe().transpose()
df_std = standarize_data(df1,df1_stats)
```

```python
# Pearson

df_pearson = df_std.corr(method='pearson')
df_pearson.iloc[-1]

# >0.3 
# % Black                           -0.313285    
# % Hispanic or Latino              -0.400971
# % married                          0.411748
# % age >=60                         0.414996
# % <highschool                     -0.431737
# median income                      0.597384
# % below poverty line              -0.401606
# % food stamp/SNAP                 -0.424104
# median value units built           0.519390
# % renter-occupied housing units   -0.389153
# population density                -0.330216
# fitness                            0.597171
```

```python
# Pearson and p-value

col = list(df_std)

for vari in col:
    if vari != 'park':
        coffi = scipy.stats.pearsonr(df_std[vari], df_std['park'])[0].round(4)
        p_value = scipy.stats.pearsonr(df_std[vari], df_std['park'])[1]
        if p_value < 0.001:
            print(vari + " | " + str(coffi) + "***") 
        if p_value < 0.01 and p_value >= 0.001:
            print(vari + " | " + str(coffi) + "**") 
        if p_value < 0.05 and p_value >= 0.01:
            print(vari + " | " + str(coffi) + "*") 
        if p_value >= 0.05:
            print(vari + " | " + str(coffi)) 
```

```python
# spearman

df_spearman = df_std.corr(method='spearman')
df_spearman.iloc[-1]

# >0.3 
# % Black                           -0.338110
# % Hispanic or Latino              -0.419286
# % married                          0.352966
# % age 18-29                       -0.338636
# % age >=60                         0.398815
# % <highschool                     -0.483362
# median income                      0.553575
# % below poverty line              -0.496012
# % food stamp/SNAP                 -0.509999
# median value units built           0.477849
# % renter-occupied housing units   -0.368519
# population density                -0.422946
# fitness                            0.515639
```
